```python
# tiny shakespeare dataset
# !wget https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt
with open('input.txt', 'r', encoding='utf-8') as f:
    text = f.read()
data = text[:1000] # first 1000 characters
print(data[:100]) # print first 100 characters
```

```python
import tiktoken
enc = tiktoken.get_encoding('gpt2')
tokens = enc.encode(data)
```

Take long sequences of data
View them as 2-dimentional, make sure to load 1 additional token for the last target


```python
import torch
buf = torch.tensor(tokens[:24 +1]) # (B, T+1) to offset things and view them 
x = buf[:-1].view(4,6) # input to the transformer
y = buf[1:].view(4,6) # labels
print(x)
print(y)
```
